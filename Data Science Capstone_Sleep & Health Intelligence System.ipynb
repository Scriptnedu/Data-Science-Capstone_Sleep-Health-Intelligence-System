#### Feature Engineering and grouping 

    Based on the American Heart Association (AHA) guidelines, blood pressure (BP) is categorized as normal (<120/<80 mmHg), elevated (120-129/<80 mmHg), Stage 1 Hypertension (130-139/80-89 mmHg), or Stage 2 Hypertension ($\ge$140/$\ge$90 mmHg). A reading over 180/120 mmHg is classified as a hypertensive crisis requiring immediate care. 

    Since raw numbers might be difficult to interpret, it is necessary to categorize the blood pressure into Normal, Elevated and High in terms of health standards, this allows us assess cardiovasular risks levels rather than just raw numbers.

```python
# Using a function to categorize the Blood pressure
def BP_classification(df):
    if df['Systolic BP'] < 120 and df['Diastolic BP'] < 80:
        return 'Normal'
    elif 120 <= df['Systolic BP'] < 130 and df['Diastolic BP'] < 80:
        return 'Elevated'
    else:
        return 'High'
    
df['BP Risk'] = df.apply( BP_classification, axis = 1)
```

#### Sleep Quality Grouping

Based on scientific, medical, and clinical sleep research, sleep quality is often measured using subjective tool. A commonly used 1â€“10 scale, single-item, self-administered "Sleep Quality Scale" (SQS) is used to categorize sleep from 0 (terrible) to 10 (excellent). According to these studies, the groupings are: 
1. 7â€“10 (Excellent/Good Sleep Quality): The individual feels fully rested, refreshed, and energized upon waking, with minimal disruptions
2. 4â€“6 (Fair/Moderate Sleep Quality): Sleep is moderate, with occasional difficulty falling asleep or waking up at night. The individual may feel somewhat tired or experience mild daytime fatigue.
3. 1â€“3 (Poor/Bad Sleep Quality): Sleep is poor, with consistent trouble falling asleep, frequent wakefulness, or early morning awakening. This is often associated with high fatigue, daytime dysfunction, and, according to studies of insomnia patients, low scores on the 10-point scale. 

Hence, it is necessary for us to group the sleep quality into: poor, fair and good because it simplifies the interpretation of subjective wellness

```python
# function for sleep quality grouping
def sleep_quality_group(df):
    if 1<= df['Quality of Sleep'] <= 3:
        return 'Poor'
    elif 4 <= df['Quality of Sleep'] <= 6:
        return 'Fair'
    elif 7 <= df['Quality of Sleep'] <= 10:
        return 'Good'
    
df['Quality_of_Sleep_group'] = df.apply(sleep_quality_group, axis = 1)
```

#### Sleep Duration Grouping

    Based on scientific medical research and recommendations from organizations like the American Academy of Sleep Medicine (AASM) and the National Sleep Foundation (NSF), adult sleep duration can be categorized as follows: 
 
    1. Insufficient (<7 hours): Linked to poor mental health, weight gain, depression, and reduced alertness.
    2. Adequate (7-9 hours): Promotes optimal health, mental functioning, and physical well-being.
    3. Excessive (>9 hours): While sometimes needed for recovery from debt or illness, chronically sleeping >9 hours is associated with increased risk of morbidity and mortality. 

    This scale (1-10 hours) is for habitual nightly sleep in healthy adults (18-64 years). Categorizing sleep duration helps interpret adequacy from a health perspective

```python
# function for sleep duration grouping
def sleep_duration_grouping(df):
    if df['Sleep Duration'] < 7:
        return 'Insufficient'
    elif 7 <= df['Sleep Duration'] <=9:
        return 'Adequate'
    elif df['Sleep Duration'] > 9:
        return 'Excessive'
    
df['Sleep_duration_group'] = df.apply(sleep_duration_grouping, axis =1)
```

    From the above, it can be observed that most professionals captured in this dataset had Adequate amount of sleep and none had Excessive sleep duration.

#### Physical Activity Grouping

    Physical activity levels are generally categorized based on daily or weekly minutes of Moderate-to-Vigorous Physical Activity (MVPA). The most common framework for adults (18-64 years) is based on guidelines from the CDC and WHO, which recommend 150â€“300 minutes of moderate-intensity or 75â€“150 minutes of vigorous-intensity aerobic activity per week, translating to roughly 20-45 minutes per day. 

    For this analysis, we will assume that the data for physical activity was collected in a moderate-intensity state. Hence, we will categorize physical activity levels as follows:
    1. Low: < 20 mins/day
    2. moderate: 20 - 40 mins/day
    3. high: > 40 mins/day

```python
# function for physical activity level grouping
def activity_level_group(df):
    if df['Physical Activity Level'] < 20:
        return 'Low'
    elif 20 <= df['Physical Activity Level'] <= 40:
        return 'Moderate'
    elif df['Physical Activity Level'] > 40:
        return 'High'
    
df['Activity_Level_Group'] = df.apply(activity_level_group, axis = 1)
```
